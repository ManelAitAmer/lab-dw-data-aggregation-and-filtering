```python
# your code goes here
import pandas as pd
```

```python
link='https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis.csv'
```

```python
insurance_data=pd.read_csv(link)
insurance_data.head()
```

```python
insurance_data.info()
```

```python
insurance_data=insurance_data.drop(columns=['Unnamed: 0'])
```

```python
insurance_data.shape
```

```python
insurance_data.isnull().sum()
```

```python
insurance_data.describe()
```

```python
insurance_data.duplicated().sum()
```

```python
insurance_data=insurance_data.drop_duplicates()
```

```python
len(insurance_data)
```

```python
insurance_data.drop_duplicates()
```

```python
insurance_data.columns=insurance_data.columns.str.lower().str.replace(" ","_")
```

```python
insurance_data.head()
```

```python
new_customer =insurance_data[(insurance_data['total_claim_amount'] > 1000) & (insurance_data['response'] == 'Yes')]
new_customer
```

```python
insurance_data_yes = insurance_data[insurance_data['response'] == 'Yes']
```

```python
average_total_claim_amount=insurance_data_yes.groupby(['policy_type','gender'])['total_claim_amount'].mean().reset_index()
average_total_claim_amount
```

```python
total_customer_plicies_by_state = insurance_data.groupby('state').size()
total_customer_plicies_by_state 
```

```python
state_filter = total_customer_plicies_by_state[total_customer_plicies_by_state > 500]
state_filter
```

```python
customer_lifetime_value_describe=insurance_data.groupby(['education','gender'])['customer_lifetime_value'].agg(['mean', 'max', "min"])
customer_lifetime_value_describe
```

```python
customer_lifetime_value_describe.groupby('gender').get_group('F').head()
```

```python
customer_lifetime_value_describe.groupby('gender').get_group('M').head()
```

### Conclusion: 

##### Gender : 
Across all education levels, there are small differences in the mean customer lifetime value between genders. For example, females with a Bachelor's and High School or Below  tend to have slightly higher mean customer lifetime values compared to their male. However, males with the rest of educations have higher mean customer lifetime value compared to females with the same education level.


##### Range of Values: 
The maximum and minimum customer lifetime values also vary across education levels and genders. For example, males with a High School or Below education level have the highest maximum customer lifetime value, while females with a Bachelor's degree have the highest maximum customer lifetime value. Similarly, females with a High School or Below education level have the highest minimum customer lifetime value, while males with a Doctorate degree have the lowest minimum customer lifetime value.
